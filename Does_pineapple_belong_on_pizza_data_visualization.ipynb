```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {'Pineapple':82, 'Pizza':84, 'Pineapple on pizza':53}
value1 = list(data.keys())
value2 = list(data.values())
fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(11)

plt.style.use('seaborn-dark-palette')
plt.bar(value1, value2, color='y');
plt.xlabel('People that like pizza and pineapple, separately and together', fontsize=13);
plt.ylabel('Count', fontsize=13);
plt.title('People that enjoy pineapple on pizza in britian', fontsize=14);
plt.xticks(fontsize=11);
plt.yticks(fontsize=11);
```
